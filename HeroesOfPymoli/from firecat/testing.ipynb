```python
# testing averages with groupby
import pandas as pd

# You can also replace the below file path to the URL of the file
df = pd.read_excel(r"Sample-Sales-Data.xlsx", sheet_name="Sheet")
```

```python
df["Total Amount"] = df["Quantity"] * df["Price Per Unit"]
df
```

```python
#df.groupby(["Salesman","Item Desc"])["Total Amount"].sum()
#age_group = ages_df.groupby('Age Group')
#df.groupby(["Salesman","Item Desc"])
# --> this works foop_df = df.groupby(["Salesman", "Item Desc"]).agg({"Total Amount" : "sum"})
#foop_df
foop_df = df.groupby(["Salesman"])\
.agg({"Total Amount" : "sum"})[["Total Amount"]]\
.apply(lambda x: 100*x/x.sum())\
.sort_values(by="Total Amount", ascending=False)
foop_df

```
